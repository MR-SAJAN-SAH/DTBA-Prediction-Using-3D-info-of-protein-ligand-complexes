```python
import py3Dmol

def show_3d(pdb_path):
    with open(pdb_path, 'r') as f:
        pdb_block = f.read()

    view = py3Dmol.view(width=600, height=500)
    view.addModel(pdb_block, 'mol2')
    view.setStyle({'cartoon': {'color': 'spectrum'}})
    view.zoomTo()
    return view.show()
show_3d("pdbbind/data/1a07/1a07_ligand.mol2")
```

```python
import py3Dmol

def show_3d(pdb_path):
    with open(pdb_path, 'r') as f:
        pdb_block = f.read()

    view = py3Dmol.view(width=600, height=500)
    view.addModel(pdb_block, 'pdb')
    view.setStyle({'cartoon': {'color': 'spectrum'}})
    view.zoomTo()
    return view.show()

show_3d("pdbbind/data/1a07/1a07_protein.pdb")
```

```python
import py3Dmol

# load the ligand (.mol2) and protein (.pdb)
ligand_file = 'pdbbind/data/1a07/1a07_ligand.mol2'
protein_file = 'pdbbind/data/1a07/1a07_protein.pdb'

with open(protein_file, 'r') as f:
    protein = f.read()

with open(ligand_file, 'r') as f:
    ligand = f.read()

# create the 3D viewer
view = py3Dmol.view(width=600, height=500)
view.addModel(protein, 'pdb')
view.setStyle({'cartoon': {'color': 'spectrum'}})  # protein as ribbon

view.addModel(ligand, 'mol2')
view.setStyle({'model': 1}, {'stick': {'colorscheme': 'cyanCarbon'}})  # ligand as sticks

view.zoomTo()
view.show()
```
